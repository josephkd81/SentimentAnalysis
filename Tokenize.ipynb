```python
f = open('to_token.txt', 'r')
lines = f.readlines()
texts = []

for line in lines:
    texts.append(line)

f.close()
```

### [tokenizer in Keras](https://keras.io/preprocessing/text/)

```python
from keras.preprocessing.text import Tokenizer

tk = Tokenizer(num_words=None, char_level=True, oov_token='_UNK')

tk.fit_on_texts(texts)
```

```python
vocab_token = tk.word_index
```

```python
print(vocab_token)
```

```python
vocab_token_list = list(vocab_token)
```

```python
len(vocab_token_list)
```

```python
vocab_token_list.append('_PAD')
```

```python
len(vocab_token_list)
```

```python
idx = []
idx_ = -1

for i in range(len(vocab_token_list)):
    idx_ += 1
    idx.append(idx_)
    
#df_idx = pd.DataFrame(idx, columns=['token_idx'])

#df_vocab_token = pd.merge(data, df_idx, columns=['char', 'idx'])
```

```python
len(idx)
```

```python
import pandas as pd

df = pd.DataFrame(vocab_token_list, columns=['char'])
df_idx = pd.DataFrame(idx, columns=['idx'])

df_vocab_token = pd.concat([df, df_idx], axis=1)
```

```python
df_vocab_token
```

```python
import csv

df_vocab_token.to_csv(r'Vocab.token.txt', header=None, index=None, sep='\t', mode='a')
```

- _PAD 와 _UNK 는 수작업으로 각각 1, 0 으로 바꿔주었음.
